```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Telco-Customer-Churn.csv')
```

```python
df.sample(5)
df.info()
```

```python
df.drop(['customerID','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','PaperlessBilling','PaymentMethod'], axis=1, inplace=True)
```

```python
df.sample(5)
```

```python
df.isnull().sum()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Bar Graph
ax1.set_title('Bar Graph')
sns.countplot(x='gender', data=df, ax=ax1)

# Pie Chart
ax2.set_title('Pie Chart')
df['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=ax2)
plt.show()
```
